```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
sns.set()
```

## Load the data 

```python
penguins = sns.load_dataset('penguins')
```

```python
penguins
```

## Analyze dataset 

### 1.The relation between the penguins flipper length according to the species.

```python
sns.relplot(x='species', y='flipper_length_mm', hue='sex' ,kind='line' ,data=penguins) 
```

## Verbalizing

The plot above shows the that penguins's flipper length diversed from specie to specie  where the Gentoo have the longest filpper and the Adelie have the shortest.
Also shows that the males of penguins ahve longer flipper than the females
